```python
# Записываем в csv
import csv

students = [
    ["Greg", "Dean", 70, 80, 90, "Good job, Greg"],
    ["Wirt", "Wood", 80, 80.2, 80, "Nicely done"]
]

with open("module3/example.csv", 'a') as f:
    writer = csv.writer(f)
    for student in students:
        writer.writerow(student) #  построчная запись в csv
```

```python
# Записываем в csv
import csv

students = [
    ["Greg", "Dean", 70, 80, 90, "Good job, Greg"],
    ["Wirt", "Wood", 80, 80.2, 80, "Nicely done"]
]

with open("module3/example.csv", 'a') as f:
    writer = csv.writer(f, quoting=csv.QUOTE_ALL) #  Помещаем в ковычки все данные которые записываем 
    writer.writerows(students)
```

```python
import csv

with open("module3/Crimes.csv") as f:
    reader = csv.reader(f)
    crime = []
    count = {}
    for row in reader:
        if ('2015' in row[2]):
            crime.append(row[5])
    for i in crime:
        if i not in count:
            count.update()
```
